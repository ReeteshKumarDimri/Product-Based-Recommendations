# Data Cleaning

```python
reviewer_id_split = reviewers['user_id'].str.split(',', expand=False)
reviewer_id_exp = reviewer_id_split.explode()
reviewer_id_clean = reviewer_id_exp.reset_index(drop=True)
reviewer_id_clean
```

```python
reviewer_name_split = reviewers['user_name'].str.split(',', expand=False)
review_name_exp = reviewer_name_split.explode()
reviewer_name_clean = review_name_exp.reset_index(drop=True)
reviewer_name_clean
```

```python
df21 = pd.DataFrame(data=reviewer_id_clean)
df22 = pd.DataFrame(data=reviewer_name_clean)
df2 = pd.merge(df21, df22, left_index=True, right_index=True)
df2=pd.DataFrame(df2)
df2.head(10)
len(df2)
```

```python
df_main=pd.DataFrame()
df_main
```

```python
start=0
t=0
l=[]
for i in range(len(data)):
    for j in range(start,start+8):
        if (j<len(df2)):
            df_temp=pd.DataFrame({'product_id':data.loc[i]['product_id'],'product_name':data.loc[i]['product_name'],'category':data.loc[i]['category'],'discounted_price':data.loc[i]['discounted_price'],'actual_price':data.loc[i]['actual_price'],'discount_percentage':data.loc[i]['discount_percentage'],'rating':data.loc[i]['rating'],'rating_count':data.loc[i]['rating_count'],'about_product':data.loc[i]['about_product'],'user_id':df2.loc[j]['user_id'],'user_name':df2.loc[j]['user_name'],'review_id':data.loc[i]['review_id'],'review_title':data.loc[i]['review_title'],'review_content':data.loc[i]['review_content'],'img_link':data.loc[i]['img_link'],'product_link':data.loc[i]['product_link']},index=[t])
            l.append(df_temp)
            t=t+1
    start=start+8
            
            
df_main=pd.concat(l,ignore_index = True)      
```

# Popularity Based Recommendation

```python
avg_rating_df=df_main.groupby('product_name').mean(numeric_only=True)['rating'].reset_index()
avg_rating_df.rename(columns={'rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df=avg_rating_df[avg_rating_df['avg_rating']>=3].sort_values('avg_rating',ascending=False).head(100)
popular_df.head()
```

```python
popular_df=popular_df.merge(df_main,on='product_name').drop_duplicates('product_name')[['product_id','product_name','img_link','product_link','avg_rating','discounted_price']]
popular_df.head(5)
```

```python
popular_df.to_csv('popular_dataset.csv')
```

# Collaborative Based Recommendation

```python
pt=df_main.pivot_table(index='product_name',columns='user_id',values='rating')
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
```

```python
def recommend(product_name):
    index=np.where(pt.index==product_name)[0][0]
    similar_items=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    data=[]
    for i in similar_items:
        item=[]
        temp_df=df_main[df_main['product_name']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('product_name')['product_name'].values))
        item.extend(list(temp_df.drop_duplicates('product_name')['img_link'].values))
        item.extend(list(temp_df.drop_duplicates('product_name')['discounted_price'].values))
        item.extend(list(temp_df.drop_duplicates('product_name')['rating'].values))
        item.extend(list(temp_df.drop_duplicates('product_name')['product_id'].values))
        data.append(item)
    return data
```

```python
recommend('Spigen EZ Fit Tempered Glass Screen Protector for iPhone 14 Pro Max - 2 Pack (Sensor Protection)')
```
